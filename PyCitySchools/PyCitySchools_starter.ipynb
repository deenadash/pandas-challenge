```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calcualte the total number of schools
total_number_schools = len(school_data)

# calculate the total number of students;
total_number_students = len(student_data)


# calculate the total budget
total_budget = school_data["budget"].sum()

# calculate the average maths scores
average_maths_score = student_data['maths_score'].mean()


# calculate the average reading scores
average_reading_score = student_data['reading_score'].mean()

# Percentage of student with passing math_score (=,> 50)
student_passing_math = student_data["maths_score"] >= 50
percent_passing_math = ((student_passing_math).mean())*100

# Percentage of student with passing reading_score (=,> 50)
student_passing_reading = student_data["reading_score"] >= 50
percent_passing_reading = ((student_passing_reading).mean())*100

# Percentage of student who passed maths and readinge (% overall passing)

overall_passing = len(student_data[(student_data["maths_score"] >= 50) & (student_data["reading_score"] >= 50)])/total_number_students
percentage_overall_passing = overall_passing * 100

#dataframe to hold the results
area_summary = pd.DataFrame({"Total Schools": [total_number_schools],
                             "Total Students": [total_number_students],
                             "Total Budget": [total_budget],
                             "Average Math Score": [average_maths_score],
                             "Average Reading Score": [average_reading_score],
                             "% Passing Maths": [percent_passing_math],
                             "% Passing Reading": [percent_passing_reading],
                             "% Overall Passing": [percentage_overall_passing]})

#formating for cleaner outlook
area_summary["Total Budget"] = area_summary["Total Budget"].map("${:,.2f}".format)
area_summary["Total Students"] = area_summary["Total Students"].map("{:,}".format)
area_summary["Average Math Score"] = area_summary["Average Math Score"].map("{:.2f}".format)
area_summary["Average Reading Score"] = area_summary["Average Reading Score"].map("{:.2f}".format)
area_summary["% Passing Maths"] = area_summary["% Passing Maths"].map("{:.2f}%".format)
area_summary["% Passing Reading"] = area_summary["% Passing Reading"].map("{:.2f}%".format)
area_summary["% Overall Passing"] = area_summary["% Overall Passing"].map("{:.2f}%".format)
area_summary
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
school_grp = school_data_complete.groupby(["school_name"])

school_type = school_grp["type"].first()

total_students = school_grp["Student ID"].count()

total_school_budget = school_grp["budget"].first()

#calculation:
student_per_budget = total_school_budget/total_students

average_maths_score = school_grp["maths_score"].mean()


average_reading_score = school_grp["reading_score"].mean()

percent_passing_maths = school_data_complete[school_data_complete["maths_score"] >= 50].groupby(["school_name"]).count()["Student ID"] / total_students * 100

percent_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 50].groupby(["school_name"]).count()["Student ID"] / total_students * 100

percent_overall_passing = school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)].groupby(["school_name"]).count()["Student ID"] / total_students * 100

#dataframe to hold the results
school_summary = pd.DataFrame({"School Type": school_type,
                             "Total Students": total_students,
                             "Total School Budget": total_school_budget,
                             "Per Student Budget": student_per_budget,
                             "Average Math Score": average_maths_score,
                             "Average Reading Score": average_reading_score,
                             "% Passing Maths": percent_passing_math,
                             "% Passing Reading": percent_passing_reading,
                             "% Overall Passing": percent_overall_passing})

#formating for cleaner outlook
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)
school_summary["Total Students"] = school_summary["Total Students"].map("{:,}".format)
school_summary["Average Math Score"] = school_summary["Average Math Score"].map("{:.2f}".format)
school_summary["Average Reading Score"] = school_summary["Average Reading Score"].map("{:.2f}".format)
school_summary["% Passing Maths"] = school_summary["% Passing Maths"].map("{:.2f}%".format)
school_summary["% Passing Reading"] = school_summary["% Passing Reading"].map("{:.2f}%".format)
school_summary["% Overall Passing"] = school_summary["% Overall Passing"].map("{:.2f}%".format)
school_summary
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
top_school_performance = school_summary.sort_values(by="% Overall Passing", ascending=False)
top_school_performance.head(5)
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
IX_year = student_data[student_data["year"] == 9]
IX_year_maths_scores = IX_year.groupby("school_name")["maths_score"].mean()

X_year = student_data[student_data["year"] == 10]
X_year_maths_scores = X_year.groupby("school_name")["maths_score"].mean()

XI_year = student_data[student_data["year"] == 11]
XI_year_maths_scores = XI_year.groupby("school_name")["maths_score"].mean()

XII_year = student_data[student_data["year"] == 12]
XII_year_maths_scores = XI_year.groupby("school_name")["maths_score"].mean()

#dataframe to store results:
mathsscores_byyear = pd.DataFrame({"YEAR 9" : IX_year_maths_scores,
                                   "YEAR 10": X_year_maths_scores,
                                   "YEAR 11": XI_year_maths_scores,
                                   "YEAR 12": XII_year_maths_scores})

#Round the values to 2 decimal palce:
mathsscores_byyear = mathsscores_byyear.round(2)
mathsscores_byyear
```

## Reading Score by Year

* Perform the same operations as above for reading scores

```python
IX_year = student_data[student_data["year"] == 9]
IX_year_reading_scores = IX_year.groupby("school_name")["reading_score"].mean()

X_year = student_data[student_data["year"] == 10]
X_year_reading_scores = X_year.groupby("school_name")["reading_score"].mean()

XI_year = student_data[student_data["year"] == 11]
XI_year_reading_scores = XI_year.groupby("school_name")["reading_score"].mean()

XII_year = student_data[student_data["year"] == 12]
XII_year_reading_scores = XI_year.groupby("school_name")["reading_score"].mean()

#Dataframe to store the results:
readingscores_byyear = pd.DataFrame({"YEAR 9" : IX_year_reading_scores,
                                   "YEAR 10": X_year_reading_scores,
                                   "YEAR 11": XI_year_reading_scores,
                                   "YEAR 12": XII_year_reading_scores})

readingscores_byyear = readingscores_byyear.round(2)
readingscores_byyear
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
school_data_complete["Per Student Budget"] = school_data_complete["budget"] / school_data_complete["size"]

#create bins
bins = [0, 585, 630, 645, 680]
#labels for bins:
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

#slicing:
school_data_complete["Spending Ranges (Per Student)"] = pd.cut(school_data_complete["Per Student Budget"], bins, labels=group_names)

#calculation:
avg_maths_score = school_data_complete.groupby(["Spending Ranges (Per Student)"]).mean()["maths_score"]
avg_reading_score = school_data_complete.groupby(["Spending Ranges (Per Student)"]).mean()["reading_score"]
passing_math = school_data_complete[school_data_complete["maths_score"] >= 50].groupby(["Spending Ranges (Per Student)"]).count()["Student ID"] / school_data_complete.groupby(["Spending Ranges (Per Student)"]).count()["Student ID"] * 100
passing_reading = school_data_complete[school_data_complete["reading_score"] >= 50].groupby(["Spending Ranges (Per Student)"]).count()["Student ID"] / school_data_complete.groupby(["Spending Ranges (Per Student)"]).count()["Student ID"] * 100
overall_passing = school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)].groupby(["Spending Ranges (Per Student)"]).count()["Student ID"] / school_data_complete.groupby(["Spending Ranges (Per Student)"]).count()["Student ID"] * 100

#storing results in dataframe:
school_spending = pd.DataFrame({
    "Average Math Score": avg_maths_score,
    "Average Reading Score": avg_reading_score,
    "% Passing Math": passing_math,
    "% Passing Reading": passing_reading,
    "Overall Passing Rate": overall_passing
})                    
 
 #Rounding values in 2 decimal place:                           
school_spending = school_spending.round(2)
school_spending
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
bins = [0, 1000, 2000, 5000]
group_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

#binning together
school_data_complete["School Size"] = pd.cut(school_data_complete["size"], bins, labels=group_labels)

#calculation:
avg_maths_score = school_data_complete.groupby(["School Size"]).mean()["maths_score"]
avg_reading_score = school_data_complete.groupby(["School Size"]).mean()["reading_score"]
percent_passing_math = school_data_complete[school_data_complete["maths_score"] >= 50].groupby(["School Size"]).count()["Student ID"] / school_data_complete.groupby(["School Size"]).count()["Student ID"] * 100
percent_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 50].groupby(["School Size"]).count()["Student ID"] / school_data_complete.groupby(["School Size"]).count()["Student ID"] * 100
percent_overall_passing = school_data_complete[(school_data_complete["maths_score"] >= 50) & (school_data_complete["reading_score"] >= 50)].groupby(["School Size"]).count()["Student ID"] / school_data_complete.groupby(["School Size"]).count()["Student ID"] * 100


#dataframe to store result:
school_size = pd.DataFrame({
    "Average Math Score": avg_maths_score,
    "Average Reading Score": avg_reading_score,
    "% Passing Math": percent_passing_math,
    "% Passing Reading": percent_passing_reading,
    "Overall Passing Rate": percent_overall_passing
})            
      
 # Rounding values in two decimal places:                         
school_size = school_size.round(2)   
school_size
```

## Scores by School Type

* Perform the same operations as above, based on school type

```python
school_type_grouped = school_data_complete.groupby("type")

#calculation:
average_maths_score = school_type_grouped['maths_score'].mean()
average_reading_score = school_type_grouped['reading_score'].mean()


percent_passing_math = school_data_complete[school_data_complete["maths_score"] >= 50].groupby(["type"]).size()/school_type_grouped.size() * 100
percent_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 50].groupby(["type"]).size()/school_type_grouped.size() * 100

overall_passing = (percent_passing_math + percent_passing_reading)/2

#dataframe to store result
school_scores = pd.DataFrame({"Average Math Score": average_maths_score,
                              "Average Reading Score": average_reading_score,
                              "% Passing Math": percent_passing_math,
                              "% Passing Reading": percent_passing_reading,
                              "Overall Passing Rate": overall_passing})
school_scores = school_scores.round(4)
school_scores
```
